**7. Standardize the data set.**

```r
# calculate the pre-process parameters from the dataset
preprocessParams <- preProcess(d[,1:11], method=c("center", "scale"))
# summarize transform parameters
print(preprocessParams)
# transform the dataset using the parameters
transformed <- predict(preprocessParams, d[,1:11])
# summarize the transformed dataset
summary(transformed)
```

**8. Perform PCA on the standardized data capturing 80% variance.**

```r
# calculate the pre-process parameters from the dataset
preprocessParams <- preProcess(d, method=c("center", "scale", "pca"),thresh = 0.80)
# summarize transform parameters
print(preprocessParams)
# transform the dataset using the parameters
transformed <- predict(preprocessParams, d)
# summarize the transformed dataset
summary(transformed)
```
